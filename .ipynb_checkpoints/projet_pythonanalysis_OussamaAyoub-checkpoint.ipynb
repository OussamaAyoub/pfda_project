### Multi-Class Model and Representation

```python
from sklearn.metrics import roc_curve, auc
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import LinearSVC
from sklearn.preprocessing import label_binarize
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Changements des différents dataset pour pouvoir créer des courbes ROC pour chaque label
X=df.drop("LABEL",axis="columns") #données en entrée
Y=label_binarize(df["LABEL"], classes=[0,1,2,3,4,5,6])# Création de vecteurs avec un 1 pour la bonne classe
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.15)# 85% de train, 15% de test
X_train,X_validation,Y_train,_Y_validation=train_test_split(X_train,Y_train,test_size=0.03) # 3% de validation sur l'ensemble des
#valeurs train
X_train.shape,X_validation.shape, X_test.shape
n_classes=Y.shape[1]
```

```python
# classifieur
clf = OneVsRestClassifier(LinearSVC())
Y_score = clf.fit(X_train, Y_train).decision_function(X_test) 

# Calcul des courbes ROC et les aires pour chaque classe
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(Y_test[:, i], Y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
```

```python
plt.figure()
for i in range(n_classes): # pour chaque classe, on récupère les taux faux positifs et vrais positifs et on affiche la courbe ROC
    plt.plot(fpr[i], tpr[i], label='ROC curve (area = %0.2f)' % roc_auc[i])
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic '+levels[i])
    plt.legend(loc="lower right")
    plt.show()
```
